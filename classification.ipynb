# bace

```python
import pandas as pd

bace = pd.read_csv('Project/SA_MIC_reg/cv_report_feature_selectors.csv')
bace
```

```python
from ProQSAR.qsar import ProQSAR
from ProQSAR.Config.config import Config
from ProQSAR.Featurizer.feature_generator import FeatureGenerator
import matplotlib
matplotlib.use("Agg")

from rdkit import RDLogger 
RDLogger.DisableLog('rdApp.*') 

import warnings
warnings.filterwarnings('ignore')

smiles_col, id_col, activity_col = 'mol', 'CID', 'Class'
feature_type = FeatureGenerator.get_all_types()

config = Config(
    featurizer={"feature_types": feature_type},
    splitter={'test_size': 0.1, 'option': 'scaffold'},
    optimizer={'n_trials': 100}
)
qsar = ProQSAR(activity_col, id_col, smiles_col, n_jobs=8, n_splits=5, n_repeats=5, config=config, project_name="bace")
qsar.run_all(data_dev=bace, alpha=[0.05, 0.1, 0.2])
```

# bbbp

```python
import pandas as pd

bbbp = pd.read_csv('Data/classification/bbbp.csv')
bbbp
```

```python
from ProQSAR.qsar import ProQSAR
from ProQSAR.Config.config import Config
from ProQSAR.Featurizer.feature_generator import FeatureGenerator
import matplotlib
matplotlib.use("Agg")

from rdkit import RDLogger 
RDLogger.DisableLog('rdApp.*') 

import warnings
warnings.filterwarnings('ignore')

smiles_col, id_col, activity_col = 'smiles', 'name', 'p_np'
feature_type = FeatureGenerator.get_all_types()
```

```python
from ProQSAR.data_generator import DataGenerator

config = Config(
    featurizer={"feature_types": feature_type, "deactivate": True},
    splitter={'test_size': 0.1, 'option': 'scaffold'},
    optimizer={'n_trials': 100}
)
datagen = DataGenerator(activity_col, id_col, smiles_col, n_jobs=1, save_dir=None, config=config)
data = datagen.generate(bbbp)
```

```python
null_data = data[data['standardized_mol'].isnull()]
null_data
```

```python
null_record = null_data[smiles_col].to_list()
```

```python
from rdkit import Chem
for i in null_record:
    mol = Chem.rdmolfiles.MolFromSmiles(i)
    print(mol)
```

```python
bbbp_new = bbbp.set_index(smiles_col)
bbbp_new = bbbp_new.drop(index=null_record).reset_index()
bbbp_new
```

```python
bbbp_new.to_csv('Data/classification/bbbp_new.csv', index=False)
```
