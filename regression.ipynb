# freesolv

```python
import pandas as pd

freesolv = pd.read_csv('Data/regression/freesolv.csv')
freesolv
```

```python
from ProQSAR.qsar import ProQSAR
from ProQSAR.Config.config import Config
from ProQSAR.Featurizer.feature_generator import FeatureGenerator
import matplotlib
matplotlib.use("Agg")

smiles_col, id_col, activity_col = "smiles", "smiles", "y"
feature_type = FeatureGenerator.get_all_types()

config = Config(
    featurizer={"feature_types": feature_type},
    splitter={'test_size': 0.1, 'option': 'scaffold'},
    optimizer={'n_trials': 100}
)
qsar = ProQSAR(activity_col, id_col, smiles_col, n_jobs=8, n_splits=5, n_repeats=5, config=config, project_name="freesolv")
qsar.run_all(data_dev=freesolv, alpha=[0.05, 0.1, 0.2])
```
