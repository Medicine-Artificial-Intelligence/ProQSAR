```python
import pandas as pd

# Replace data path
train = pd.read_csv('Data/admet_group/hia_hou/train_val.csv')
test = pd.read_csv('Data/admet_group/hia_hou/test.csv')

train
```

```python
from ProQSAR.qsar import ProQSAR
from ProQSAR.Config.config import Config
from ProQSAR.Featurizer.feature_generator import FeatureGenerator
import matplotlib
matplotlib.use("Agg")

feature_type = FeatureGenerator.get_all_types()

# Change the column name according to each data
smiles_col, id_col, activity_col = "Drug", "Drug_ID", "Y"

config = Config(
    featurizer={"feature_types": feature_type},
    splitter={'deactivate': True}, # deactivate Splitter
    optimizer={'n_trials': 100,} # may tap truoc optimize nhiu trials doi lai cho giong gium chi nha
)
# change project name
qsar = ProQSAR(activity_col, id_col, smiles_col, scoring_target='roc_auc', n_jobs=4, n_splits=5, n_repeats=5, config=config, keep_all_test=True, keep_all_pred=True, project_name="hia_hou")
qsar.run_all(data_dev=train, data_test=test, alpha=[0.05, 0.1, 0.2])
```
